Masking And Vectorization

```python
m=np.arange(1,21).reshape(5,-1)
print(m)
```

```python
print(m)
```

```python
mask=m>=10
print(mask)
```

```python
mask=m%2==0
print(mask)
```

```python
print(m[mask])
```

```python
import math
vefun=np.vectorize(math.sin)
vefun(m)
```

```python
def myfunc(a,b):
    if a > b :
        return a - b
    else:
        return a + b
```

```python
vefun=np.vectorize(myfunc)
vefun([1,2,3,4],2)
```

**Stacking and Special Array**

```python
A=np.arange(1,5).reshape(2,2)
B=np.arange(11,19).reshape(2,4)
print(A)
print(B)
```

```python
np.hstack((A,B))
```

```python
np.vstack((A,B.T))
```
