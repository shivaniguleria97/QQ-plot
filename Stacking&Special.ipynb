```python
import numpy as np
```

```python
A=np.arange(1,5).reshape(2,2)
B=np.arange(11,19).reshape(2,4)
print(A)
print(B)
```

```python
np.hstack((A,B))
```

```python
np.vstack((A,B.T))
```
