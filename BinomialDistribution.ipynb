```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
n_patient=1000
p_sucess=0.9
dist_binomial=stats.binom(n_patient,p_sucess)
x_values=np.arange(500,n_patient+1,step=5)
y_values=dist_binomial.pmf(x_values)
plt.stem(x_values,y_values)
```
